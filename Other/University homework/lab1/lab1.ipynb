# Можно сделать через lambda x: log(x)

```python
cols = file.select_dtypes(include='number') 
data = cols.values
data_log = np.log(data, where=data > 0)
print(data)
print(data_log)
```

```python
for values in data_log:
    column = values
    print(plt.boxplot(column))
    plt.show()
    
for values in data_log:
    column = values
    print(plt.hist(column))
    plt.show()
```
