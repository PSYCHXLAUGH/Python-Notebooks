# Потоки (Thread)

Потоком управляет ядро
Поток - последовательность операций выполнениями которых занимается операционная система. Потоки создаются внутри процесса и потребляют меньше ресурсов. У каждого процесса есть 1 поток **MainThread** в котором выполняются инструкции которые были написаны в программе. Процесс в свою очередь может запускать дополнительные потоки помимо **MainThread** таким образом делать параллель между вычислениями. Все потоки в рамках одного процесса разделяют ресурсы. В частности разделяют оперативную память (в этом и плюс и минус)
- **Плюсы**
<br>

Мы можем поделить логику в несколько потоков. Разные потоки могут лечь в разные процессорные ядра. Таким образом мы можем ускорить работу.
- **Минусы**
<br>

В многопоточных программах есть **проблема**
**проблема - состояния гонки** когда время выполнения программы начинает зависеть от порядка выполнения кода в разных потоках. Например один поток может изменять какие-то данные а другой в это время может читать эти данные и он может не рассчитывать что эти же данные могут быть не изменены **другим потоком**. Эти проблемы сложно отслеживать потому что они возникают не всегда. Они возникают только в ряде случаев. Проблема в том что потоки должны выполнятся в особом порядке (в таких случаях эти проблемы начинают проявлятся) Это очень сильная проблема в многопоточных программах. 

Потоки позволяют эффективно задействовать ресурсы компьютера. Например у нас есть процессор с 4 ядрами мы можем написать программу, которая создаст 4 потока и получится так, что каждый поток ляжет на отдельное ядро и потоки будут выполнятся одновременно с учетом того, что в системе запущены какие-либо другие процессы тоже. Они имеют возможность выполнятся на разных процессорных ядрах.

Также существенный минус потоков в том, что мы не можем создавать "сколько угодно" потоков, но ядер у нас ограниченное количество и будет все работать медленно. Но параллель между логикой сделать хочется. 
Для этого и существует **асинхронность или кооперативная многозадачность** Имеет ли это отношение к процессам и потокам? Нет. Асинхронность работает в рамках одного процесса и в рамках одного потока. 

# Hyper-Threading

Разбиение каждого физического ядра на 2 виртуальных. В таком случае может проходить через каждое ядро в 2 раза больше информации. Но значит ли это что 2 виртуальных ядра = 2 физических? Нет, поскольку если 1 ядро будет сильно загружено по 1 потоку, то 2 ждет в очереди. Разные программы взаимодействуют с процессором по разному, иногда эта технология дает производительность.

Например если у процессора 6 ядер и 12 потоков, то у каждого ядра по 2 hyper-треда и операционная система будет видеть не 6 ядер а 12 ядер. 12 физических ядер лучше чем 12 с технологией hyper-threading.

# Асинхроность

Асинхронность, кооперативная многозадачность. Это не имеет отношения к процессам и потокам. Асинхронность работает в рамках одного процесса и в рамках одного потока **MainThread**. Мы пишем например веб приложение и у нас клиенты его вызывают по http. Он отвечает за 300мс **например**. Если мы выясним что происходит в течении этих 300 мс, то мы поймем что время расходуется на ожидание в основном. Мы отправляем запрос в базу данных 1, 2, 3 раз и если мы проссумируем ожидание ответа от базы данных то получится 200мс например. за 100мс у нас отрабатывает сервер и остальные 200 у нас запросы в базу данных в какие-то внешние api. В системы кеширования. Большее количество времени у нас расходуется время на ожидание и это не удобно. Сервер в это время не может выполнять другие задачи. Например у нас есть 4 процессорных ядра + Hyper-Threading 8 потоков может "одновременно" выполнятся в нашей системе с этим процессором. И кажется что есть 8 потоков которые смогут эффективно обработать запросы и когда подключатся 8 клиентов к нашему веб серверу все +- хорошо, но когда их подключается 20\30\40 то уже не все обрабатываются и чем длинее эта очередь накапливается, тем больше время ожидания и начинаются тайм-ауты, последние клиенты из этой очереди начинают отваливаться потому что по тайм-ауту не могут так долго ждать и тд. **Асинхронность** технология, которая позволяет это делать.
<br>
**В рамках 300мс в которых полностью обрабатывает наш сервис 200 мы ожидаем** соответственно 200мс мы можем потратить на другие запросы клиентов. Асинхронность решает эту проблему. У нас может быть 1 процесс, внутри него 1 поток и в рамках этого потокм мы обрабатывает пришедший http запрос потом мы понимаем что в рамках этого запроса нужно отправить запрос в базу данных и вместо того, чтобы заблокировать выполнение и просто ожидать ответа от базы данных мы обрабатываем следующий http запрос и также отправляем запрос в базу данных вместо того, чтобы просто ждать и так 3, 4, 5 клиента.Таким образом один поток может обрабатывать большинство клиентов. При этом очевидно, что мы можем запустить несколько копий(процессов) нашей программы. Например у нас 4-х ядерный процессор и мы можем запустить 4 или 8 процессов с нашей программой внутри каждой будет 1 поток, но каждый поток будет обрабатывать огромное количество клиентов одновременно, потому что он не блокирует свой поток выполнения и когда он что-то ожидает то обрабатывает других клиентов. 

Есть **event_loop()** цикл событий, есть задачи которые ожидают завершения, когда эти задачи завершаются возникает callback и этот цикл событий отслеживает эти callback'и и когда они возникают он вызывает их. В моменты когда еще callback еще не пришел он занимается выполнением других задач. event_loop() цикл который управляет задачами. Выполняет одну задачу до тех пор пока она не встает в ожидание. Как только она встает в ожидание цикл управления передается в какую-то другую задачу выполняется она потом она тоже встает в ожидание возникает callback, далее он вызывается и происходит дальнейшая обработка.

```python
class Thread:
    """A class that represents a thread of control.

    This class can be safely subclassed in a limited fashion. There are two ways
    to specify the activity: by passing a callable object to the constructor, or
    by overriding the run() method in a subclass.

    """

    _initialized = False

    def __init__(self, group=None, target=None, name=None,
                 args=(), kwargs=None, *, daemon=None):

```
<pre>
target - ссылка на функцию
name - имя потока
args - позиционные аргументы функции ( кортеж )
kwargs - аргументы ключевые ( словарь )
daemon - Является ли поток демоном или это обычный поток
</pre>

```python

import threading
from time import sleep

counter = 4

def _thread_func():
    global counter
    for x in range(counter):
        print(f"{threading.current_thread().name} Hello world!")
        sleep(counter % counter + 1)



if __name__ == '__main__': 

    # Создаем объект на наш поток
    _thread = threading.Thread(
        target=_thread_func
    )

    # Вызов метода start. Запускаем поток
    _thread.start() # _thread func
    
    # Основные методы
    
    # _thread.name - имя потока
    
    # _thread.ident - идентификатор
    
    # Pid (Process ID) - это уникальный идентификатор процесса
    # который используется операционной системой для управления процессами
    # Каждый процесс имеет свой собственный Pid.
    # Идентификатор потока (Thread ID) - это уникальный идентификатор потока внутри процесса
    # Каждый поток имеет свой собственный идентификатор потока.
    
    # _thread.demon - является ли поток демоном
    
    # _thread.is_alive() - работает ли поток в данный момент

    for x in range(counter): # main thread
        print(f"{threading.current_thread().name} Hello world!")
        sleep(counter % counter + 1)
```

#### daemon threads

Потоки-демоны (daemon threads) - это потоки, которые работают в фоновом режиме и не мешают основной работе программы. Они используются для выполнения задач, которые не требуют непрерывного взаимодействия с пользователем или другими процессами.

Когда главный поток (Main Thread) завершается, все потоки-демоны автоматически завершаются. Это происходит потому, что потоки-демоны не являются необходимыми для выполнения основной работы программы и должны быть завершены, когда программа завершается. Если бы они продолжали работать после завершения главного потока, это могло бы привести к нежелательным последствиям или ошибкам в программе.

# Lock и RLock

acquire - получение потока
release - освобождение потока




#### Lock

Класс, реализующий примитивные объекты блокировки. Как только поток получил блокировку, последующие попытки получить ее блокируются до тех пор, пока она не будет снята; любой поток может освободить его.

```python

counter = 4
locker = threading.Lock() # Lock


def _thread_func():
    global counter
    while True:
        locker.acquire()
        # данный участок кода заблокирован потоком. Пока поток не выполнит код, он не даст доступ другим потокам
        counter += 1
        sleep(1)
        print(counter)
        # поток разблокировал доступ для других потоков и теперь возьмет контроль тот поток, котоырй быстрее успеет захватить
        locker.release()
if __name__ == '__main__':  # main thread
    for thread in range(2):
        threading.Thread(target=_thread_func).start()

```
